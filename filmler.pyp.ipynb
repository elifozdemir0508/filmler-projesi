```python
directors = df["director"].value_counts().head(10)
plt.figure(figsize= (10,6),dpi=120)
plt.bar(directors.index, directors.values,color= "blue")
plt.xlabel("Yönetmenler")
plt.ylabel("Çektikleri film sayisi")
plt.tight_layout()
plt.show
```

```python
genre = df["genre"].value_counts().head(5)
plt.figure(figsize=(10,6),dpi=100)
plt.bar(genre.index, genre.values, color="red")
plt.xlabel("film türleri")
plt.ylabel("film süreleri")
plt.tight_layout()
plt.show()
```
